# intailized the State

```python
class BMI_State(TypedDict):
    weight_kg:float
    height_m:float
    bmi:float

    
    
```

# define funcation or node

```python
def calculate_bmi(state:BMI_State) ->BMI_State:
    weight=state["weight_kg"]
    height=state["height_m"]

    bmi_calcultion=weight/(height**2)
    state["bmi"]=bmi_calcultion

    return state
    

    
```

# define the graphs

```python
graph=StateGraph(BMI_State)

graph.add_node("calculate_bmi",calculate_bmi)

```

# define edges

```python
graph.add_edge(START,"calculate_bmi")
graph.add_edge("calculate_bmi",END)

```

# We Make that in One flow

```python
from langgraph.graph import START,END,StateGraph
from typing import TypedDict


class BMI_State(TypedDict):
    weight_kg:float
    height_m:float
    bmi:float


def calculate_bmi(state:BMI_State) ->BMI_State:
    weight=state["weight_kg"]
    height=state["height_m"]

    bmi_calcultion=weight/(height**2)
    state["bmi"]=bmi_calcultion

    return state

graph=StateGraph(BMI_State)

graph.add_node("calculate_bmi",calculate_bmi)


graph.add_edge(START,"calculate_bmi")
graph.add_edge("calculate_bmi",END)
 

weight=float(input("insert the the weight : "))
height=float(input("insert the the height : "))


workflow=graph.compile()
intial_state={
    "weight_kg":weight,
    "height_m":height 
}
final_state=workflow.invoke(intial_state)
print("=================================================================================")
print("                               result                                            ")
print("=================================================================================")

print(f"BMI : {final_state["bmi"]}")

```
